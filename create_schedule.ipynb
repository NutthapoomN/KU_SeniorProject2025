OPEN Excel

```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook


file_path = 'Data_set/Class_Schedule_5312.xlsx'
wb = load_workbook(file_path)
ws = wb['Sheet1']   # à¸Šà¸·à¹ˆà¸­à¸Šà¸µà¸—
```

Edit Excel

```python
def update_excel(Day, Start, End, number):
    # à¸ªà¸£à¹‰à¸²à¸‡à¸Šà¹ˆà¸§à¸‡à¹€à¸§à¸¥à¸²
    r = np.arange(Start, End + 1, 1)
    n = 0

    # à¸§à¸™à¸­à¹ˆà¸²à¸™à¸—à¸¸à¸à¹à¸–à¸§
    for row in ws.iter_rows(values_only=True):
        if n == 0:  # à¸‚à¹‰à¸²à¸¡ header à¹à¸–à¸§à¹à¸£à¸
            n += 1
            continue
        else:
            n += 1
            if row[0] == Day and row[1] in r:

                print(f'ğŸ”¹ Row {n} | Before:', row)
                ws.cell(row=n, column=3, value=number)  # à¹à¸à¹‰ Number
                print('edit row and num',[n,number])
                if number > 0:
                    ws.cell(row=n, column=4, value=1)  # AIR_STAT = 1
                else:
                    ws.cell(row=n, column=4, value=0)  # AIR_STAT = 0

    # à¸šà¸±à¸™à¸—à¸¶à¸à¹„à¸Ÿà¸¥à¹Œ
    wb.save(file_path)
    wb.close()
    print('âœ… Update successful.\n')
    return

def restart(Day):# reset Number = 0, AIR_STAT=0 all time
    for i in Day:
        update_excel(Day = i, Start=1, End = 24,number =0 )
    print('successful')
    return
    
```

```python
N=60
#Monday
update_excel(Day = 1, Start=9, End = 12,number =N )
update_excel(Day = 1, Start=13, End = 18,number =N)
#Tuseday
update_excel(Day = 2, Start=10, End = 12,number =N )
update_excel(Day = 2, Start=13, End = 18,number =N )
#Wednesday
update_excel(Day = 3, Start=9, End = 12,number =N )
update_excel(Day = 3, Start=13, End = 19,number =N )
#Thusday
update_excel(Day = 4, Start=8, End = 13,number =N )
#Friday
update_excel(Day = 5, Start=9, End = 12,number =N )
update_excel(Day = 5, Start=13, End = 16,number =N )
#Saturday
update_excel(Day = 6, Start=8, End = 16,number =N )
#Sunday
update_excel(Day = 7,Start=1,End = 24,number =0 )
```
